### Final Model

Algorithm predicting the baseline estimate for given user and item.



**If user u is unknown, then the bias bu is assumed to be zero. The same applies for item i with bi.**
> Explanation the formula :
- r^ui : the estimated rating of user u for item i.
- bui  : the baseline rating of user u for item i.
- μ : the mean of all ratings.

**Baselines can be estimated in two different ways:**

- Using Stochastic Gradient Descent (SGD).
- Using Alternating Least Squares (ALS).

> They have each parameters, for ALS.
- 'reg_i': The regularization parameter for items. Default is 10.
- 'reg_u': The regularization parameter for users. Default is 15.
- 'n_epochs': The number of iteration of the ALS procedure. Default is 10.

> for SGD.
- 'reg': The regularization parameter of the cost function that is optimized. Default is 0.02.
- 'learning_rate': The learning rate of SGD. Default is 0.005.
- 'n_epochs': The number of iteration of the SGD procedure. Default is 20.


```python
bsl_options = {'method': 'als',
               'n_epochs': 5,
               'reg_u': 12,
               'reg_i': 5
               }
algo = BaselineOnly(bsl_options=bsl_options)
```

> I will try recomendation from documentation from surpirse lib with this parameters. Let's see how the performance of our model.

```python
cross_validate(algo, data, measures=['RMSE'], cv=3, verbose=False)
```

> From cross validation score with RMSE we know there is not significant RMSE score with the default parameters, decrese only 0.001. we try with this model to predict our data. Split the data first and then we predict and see the result.

```python
trainset, testset = train_test_split(data, test_size=0.25,random_state=101)
algo.fit(trainset)
predictions = algo.test(testset)
accuracy.rmse(predictions)
```

```python
def get_Iu(uid):
    """ return the number of items rated by given user
    args: 
      uid: the id of the user
    returns: 
      the number of mvoies rated by the user
    """
    try:
        return len(trainset.ur[trainset.to_inner_uid(uid)])
    except ValueError: # user was not part of the trainset
        return 0
    
def get_Ui(mid):
    """ return number of users that have rated given item
    args:
      mid: the raw id of the item
    returns:
      the number of users that have rated the movies.
    """
    try: 
        return len(trainset.ir[trainset.to_inner_iid(mid)])
    except ValueError:
        return 0
    
result = pd.DataFrame(predictions, columns=['UserId', 'MovieId', 'Rating', 'est', 'details'])
result['Iu'] = result.UserId.apply(get_Iu)
result['Ui'] = result.MovieId.apply(get_Ui)
result['error'] = abs(result.est - result.Rating)
```
