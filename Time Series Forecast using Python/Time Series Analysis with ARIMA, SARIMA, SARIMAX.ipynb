```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings('ignore')
```

```python
from yahoo_historical import Fetcher 

dfINDF = Fetcher('INDF.jk', [2011,1,1], [2020,4,30]).getHistorical()
```

```python
dfINDF.info()
```

```python
df = dfINDF[['Date', 'Adj Close']]
```

```python
## method fillna using ffil (nan values at index 55 & 124)
df['Adj Close'].fillna(method='ffill',inplace=True)
```

```python
df['Adj Close']= round(df['Adj Close'],2)
```

```python
## convert date to format date time 
df['Date'] = pd.to_datetime(df['Date'])
```
