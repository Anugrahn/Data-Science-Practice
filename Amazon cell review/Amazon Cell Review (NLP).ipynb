#### Text cleaning
- punctuation
- lematisasi
- stop words
- regular expression

```python
## Lematisasi
from nltk.stem.wordnet import WordNetLemmatizer
## stopwords
from nltk.corpus import stopwords
## punctuation
import string
```

```python
stop = set(stopwords.words('english'))
punc = set(string.punctuation)
lemma = WordNetLemmatizer()
keywords = reviews['brand'].apply(lambda x : x.lower()).unique().tolist()
keywords.append('phone')
keywords.append('handphone')
### function clean_text : 
def clean_text(text):
    ## convert the text to lowercase
    text = text.lower()
    ## split text to the list 
    wordList = text.split()
    ## remove punctuation
    wordList = ["".join(x for x in word if (x == "'") | (x not in punc)) for word in wordList]
    ## remove stopwords 
    wordList = [word for word in wordList if word not in stop]
    ## remove keywords 
    wordList = [word for word in wordList if word not in keywords]
    ## Lemmatisation 
    wordList = [lemma.lemmatize(word) for word in wordList]
    return " ".join(wordList)
```

```python
reviews['body'] = reviews['body'].astype('str')
```

```python
reviews['clean_text'] = reviews['body'].apply(clean_text)
```

```python
from wordcloud import WordCloud 
```

```python
def word_frequency(text):
    wordList = text.split()
    # generate frequencey of word to dictionary 
    wordFreq = {word : wordList.count(word) for word in wordList}
    return wordFreq
```

```python
word_frequency('I love you, I love my mom, I love my dad, I love him, you  are my everything')
```

```python
def wordcloud_freq(word_freq,title,figure_size = (10,5)):
    wordcloud.generate_from_frequencies(word_freq)
    plt.figure(figsize=figure_size)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title(title)
    plt.show
```

```python
apple_pos = " ".join(apple[apple['positivity'] == 1]['clean_text'][0:1000])
apple_pos = word_frequency(apple_pos)
wordcloud = WordCloud(width = 5000,
                     height = 2500,
                     colormap ='inferno',
                     background_color='white')
wordcloud_freq(apple_pos, 'Most Frequent Words in the Latest 1000 Positive Reviews for Apple')
```

```python
apple_pos = " ".join(apple[apple['positivity'] == -1]['clean_text'][0:1000])
apple_pos = word_frequency(apple_pos)
wordcloud = WordCloud(width = 5000,
                     height = 2500,
                     colormap ='Blues',
                     background_color='black')
wordcloud_freq(apple_pos, 'Most Frequent Words in the Latest 1000 Negative Reviews for Apple')
```

#### Setting target 

```python
X = samsung['clean_text']
y = samsung['positivity']
```

```python
### split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=101, stratify=y)
```

```python
### Using Pipeline for modeling
## Using countvectorizer --> tokenize 
## Using TF-IDF 
## Classifier modeling
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
```

```python
pipeline = Pipeline([
    # tokenizing (counting words)
    ('bow',CountVectorizer(analyzer=clean_text)),
    # TF-IDF Scores (pembobotan)
    ('tfidf',TfidfTransformer()),
    ## modeling with MultinomialNB
    ('classification', MultinomialNB())
])
```

```python
pipeline.fit(X_train,y_train)
```

```python
predict = pipeline.predict(X_test)
```

```python
print(classification_report(y_test,predict))
```

```python
pred_proba = pipeline.predict_proba(X_test)
```

```python
# default thresshold 0.5
pred_proba 
```

```python
predict = [ 1 if prob > 0.60 else -1 for prob in pred_proba[:,1]]
```

```python
print(classification_report(y_test,predict))
```
