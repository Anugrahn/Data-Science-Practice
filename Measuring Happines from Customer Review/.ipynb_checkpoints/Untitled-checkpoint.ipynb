```python
n = 12411# number of records in file 
s = 6500 # desired sample size
listIndex = list(range(n)) # list untuk total dari data 
skip_row = sorted(random.sample(listIndex,(n-s))) # untuk melakukan 'skip' dari data yang diinginkan untuk di buang
skip_row = skip_row[1:] # agar index nya tidak ilang (0 adalah index nya bray)
fileName= 'df_not_happy.csv'
df_not_happy = pd.read_csv(fileName, skiprows=skip_row, encoding ='latin')
```

```python
n = 26521 # number of records in file 
s = 7000 # desired sample size
listIndex = list(range(n)) # list untuk total dari data 
skip_row = sorted(random.sample(listIndex,(n-s))) # untuk melakukan 'skip' dari data yang diinginkan untuk di buang
skip_row = skip_row[1:] # agar index nya tidak ilang (0 adalah index nya bray)
fileName= 'df_happy.csv'
df_happy = pd.read_csv(fileName, skiprows=skip_row, encoding ='latin')
```

```python
df_not_happy = df_not_happy.drop('Unnamed: 0',axis=1)
```

```python
df_happy = df_happy.drop('Unnamed: 0',axis=1)
```

```python
frames = [df_happy,df_not_happy]
```

```python
df = pd.concat(frames)
```
