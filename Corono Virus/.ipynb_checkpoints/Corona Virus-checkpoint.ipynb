```python
covid_3.head()
```

```python
x = covid_3.groupby(['Province/State','Country/Region']).sum()[['Confirmed','Deaths','Recovered']]
```

```python
new = pd.DataFrame(x).reset_index()
```

```python
new['Not Confirmed'] = new['Confirmed'] - new['Deaths'] - new['Recovered']
```
