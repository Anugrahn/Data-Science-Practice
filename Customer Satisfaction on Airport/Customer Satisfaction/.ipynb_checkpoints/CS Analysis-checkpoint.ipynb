### Fill Nan Data at Arrival Delay in Minutes
*Nan values filled by Departure Delay in Minutes, Type of Class*

```python
group_arr_delay2 = round(df.groupby(['Departure Delay in Minutes','Class']).mean()['Arrival Delay in Minutes'])
```

```python
def applyArrivalDelay2(x):
    arrDelay_notnull2 = df[(df['Class'] == x['Class']) & (df['Arrival Delay in Minutes'].notnull())]['Departure Delay in Minutes']
    if pd.notnull(x['Arrival Delay in Minutes']):
        return x['Arrival Delay in Minutes']
    else:
        idx = np.argmin(np.abs(arrDelay_notnull2 - x['Departure Delay in Minutes']))
        return group_arr_delay2.loc[df['Departure Delay in Minutes'][idx]][x['Class']]
```

```python
df['Arrival Delay in Minutes'] = df[['Class','Departure Delay in Minutes','Arrival Delay in Minutes']].apply(applyArrivalDelay2,axis=1)
```

## Feature Engineering
*Change Categorical columns to numerical columns*

```python
df['Class'].value_counts()
```

```python
# Label encode on column Class wih value 0-2
x['Class'] = x.Class.map({
    'Eco': 0,
    'Eco Plus' : 1,
    'Business' : 2})
```

```python
#Customer Type, Gender and Type of Travel
x = pd.get_dummies(x, columns=['Gender', 'Customer Type','Type of Travel'],drop_first=True)
```

```python
#StandardScaler for column Age
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_transformer = scaler.fit(x[['Age']])
x_scaled = scaler_transformer.transform(x[['Age']])
x_scaled = pd.DataFrame(x_scaled,columns=['Age'])
x.drop(['Age'],axis=1,inplace=True)
x = pd.concat([x,x_scaled],axis=1)
x.reset_index().drop('index',axis=1,inplace=True)
```

```python
x.head()
```

## Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.metrics import classification_report
```

```python
def class_report(x,y,model):
    X_train,X_test,y_train,y_test = train_test_split(x,y,random_state=101)
    clf = model(random_state=101)
    clf.fit(X_train,y_train)
    
    prediction_test = clf.predict(X_test)
    prediction_train = clf.predict(X_train)
    
    print('Classification Data Test')
    print(classification_report(y_test,prediction_test))
    print('========================================================')
    print('Classification Data Train')
    print(classification_report(y_train,prediction_train))
```

### Logistic Regression

```python
class_report(x,y,LogisticRegression)
```

### Decision Tree Classifier

```python
class_report(x,y,DecisionTreeClassifier)
```

### Random Forest Classifier

```python
class_report(x,y,RandomForestClassifier)
```

```python
class_report(x,y,GradientBoostingClassifier)
```
