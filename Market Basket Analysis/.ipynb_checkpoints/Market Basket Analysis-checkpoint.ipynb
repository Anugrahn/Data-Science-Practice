### Data preprocessing

```python
list_all_item = []
for i in range(0, 7501):
    list_all_item.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te.fit(list_all_item)
data = te.transform(list_all_item)
data = pd.DataFrame(data, columns = te.columns_)
```

```python
data
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

#Now, let us return the items and itemsets with at least 5% support:
frequent_itemsets= apriori(data, min_support = 0.01, use_colnames = True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
```

```python
index_to_drop = rules[rules['consequents'] == frozenset({'nan'})].index
```

```python
rules.drop(index_to_drop,inplace=True)
```

```python
index_to_drop = rules[rules['antecedents'] == frozenset({'nan'})].index
```

```python
rules.drop(index_to_drop,inplace=True)
```

```python
rules
```

```python
a = rules.consequents.values[0]
```

```python
# Function to change antecendents and consequents to list from immutable 
new_cons = []
for item in rules['consequents']:
    a = list(item)
    new_cons.append(', '.join(a))
    
```

```python
new_ante = []
for item in rules['antecedents']:
    a = list(item)
    new_ante.append(', '.join(a))
    
    
```

```python
rules['antecedents'] = new_ante
rules['consequents'] = new_cons
```

```python
rules
```

```python
index_to_drop = rules[rules['consequents'].str.contains('nan')].index
rules.drop(index_to_drop,inplace=True)
```

```python
index_to_drop= rules[rules['antecedents'].str.contains('nan')].index
rules.drop(index_to_drop,inplace=True)
```

```python
rules.reset_index(drop=True,inplace=True)
```
