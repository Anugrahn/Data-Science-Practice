```python
data_summaries = requests.get('https://indodax.com/api/summaries').json()
```

```python
pd.set_option("max_columns",75)
pd.set_option("max_rows",75)
df=pd.DataFrame(data_summaries['tickers'])
```

```python
df.loc['name'].values
```

```python
df.head(10)
```

```python
list_buy = df.loc['buy'].values
list_name = df.loc['name'].values
list_sell = df.loc['sell'].values
list_last = df.loc['last'].values
list_idr = df.loc['vol_idr'].values
df_sum = [list_buy,list_sell,list_last,list_idr]
```

```python
df_sum = pd.DataFrame(df_sum,columns=list_name,index=['Buy','Sell','Last','Vol_idr']).transpose()
```

```python
df_sum
```

```python
df_sum['Buy'] = df_sum['Buy'].astype(float)
df_sum['Sell'] = df_sum['Sell'].astype(float)
df_sum['Last'] = df_sum['Last'].astype(float)
df_sum['Vol_idr'] = df_sum['Vol_idr'].astype(float)
```
